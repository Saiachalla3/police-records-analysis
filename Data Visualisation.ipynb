# Data Visualisations after data extraction

```python
import pandas as pd
import numpy as np
import matplotlib as plt
from tqdm import tqdm
import tensorflow as tf
```

```python
df = pd.read_csv("results.csv")
```

```python
grouped = df.groupby(["Department"])
grouped = df.value_counts("Department").to_frame()
grouped.rename(columns = {0: 'Count'}).head(30)
```

```python
grouped.plot.barh(figsize = (10,10), title = "Number of records each department received and replied to")
```
